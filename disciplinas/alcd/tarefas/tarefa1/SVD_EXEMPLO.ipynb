```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.image import imread

# Carregar imagem colorida
A = imread("img16.jpg")

# Converter para tons de cinza (mdia dos canais R, G, B)
X = np.mean(A, axis=2)

# Mostrar imagem original em cinza
plt.imshow(X, cmap='gray')
plt.title("Imagem original (convertida para cinza)")
plt.axis('off')
plt.show()

# Decomposio SVD
U, S, VT = np.linalg.svd(X, full_matrices=False)

# Aproximaes com r = 5, 20, 100
for r in (5, 20, 50, 100, 200):
    Xapprox = (U[:, :r] * S[:r].reshape(1, -1)) @ VT[:r, :]
    plt.imshow(Xapprox, cmap='gray')
    plt.title(f"Aproximao com r = {r}")
    plt.axis('off')
    plt.show()

# Plot dos valores singulares e soma acumulada
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.semilogy(S)
plt.title("Valores singulares (escala log)")

plt.subplot(1, 2, 2)
plt.plot(np.cumsum(S) / np.sum(S))
plt.title("Soma acumulada dos valores singulares")

plt.show()
```
